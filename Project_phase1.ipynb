- a negative correlation means there is almost no relationship between the number of fatal injuries and the number of people who were uninjured in these incidents. if the number of fatalities goes up or down, it doesn't have a clear impact on the number of people who walk away uninjured.

- a positive correlation means that there is a clear impact between the two variables being compared. when one icrease the other increases too

```python
# fill the aircraft category column with mode

Aviation['Aircraft.Category'].fillna(Aviation['Aircraft.Category'].mode()[0], inplace=True)
```

```python
# fill the total fatal with mean
Aviation['Total.Fatal.Injuries'].fillna(Aviation['Total.Fatal.Injuries'].mean(), inplace=True)
```

```python
# fill missing data for country, location, injury severity, Aircraft damage, make, model, Engine type, purpose of flight, report status, weather with mode
Aviation['Country'].fillna(Aviation['Country'].mode()[0], inplace=True)
Aviation['Location'].fillna(Aviation['Location'].mode()[0], inplace=True)
Aviation['Injury.Severity'].fillna(Aviation['Injury.Severity'].mode()[0], inplace=True)
Aviation['Aircraft.damage'].fillna(Aviation['Aircraft.damage'].mode()[0], inplace=True)
Aviation['Make'].fillna(Aviation['Make'].mode()[0], inplace=True)
Aviation['Model'].fillna(Aviation['Model'].mode()[0], inplace=True)
Aviation['Engine.Type'].fillna(Aviation['Engine.Type'].mode()[0], inplace=True)
Aviation['Purpose.of.flight'].fillna(Aviation['Purpose.of.flight'].mode()[0], inplace=True)
Aviation['Report.Status'].fillna(Aviation['Report.Status'].mode()[0], inplace=True)
Aviation['Weather.Condition'].fillna(Aviation['Weather.Condition'].mode()[0], inplace=True)
```

```python
# check missing data
Aviation.isnull().sum()
```

```python
# fill in missing data number of engines, total serious injuries, total minor injuries, total uninjured with mean
Aviation['Number.of.Engines'].fillna(Aviation['Number.of.Engines'].mean(), inplace=True)
Aviation['Total.Serious.Injuries'].fillna(Aviation['Total.Serious.Injuries'].mean(), inplace=True)
Aviation['Total.Minor.Injuries'].fillna(Aviation['Total.Minor.Injuries'].mean(), inplace=True)
Aviation['Total.Uninjured'].fillna(Aviation['Total.Uninjured'].mean(), inplace=True)
```

```python
# drop columns Event id, with missing data
Aviation = Aviation.dropna(axis=1)
```

- Drop missing values, all data with missing data from the columns was dropped after filling in missing values in some of the colums to come up with workable and presentable data.
- After dropping we reamained with 88889 rows and 20 columns

# Visualization
## Univariate analysis - features understanding
- Plotting of feature distributions
 - Histogram, box plot


## Bivariate Analysis - features relationships.
- plotting of feature relationships
 - Scatter plot, Heatmap Correlation, Pairplot


```python
# plot a bar graph of incidents by weather condition
Aviation1['Weather.Condition'].value_counts().plot(kind='bar', title='Incidents by weather condition', figsize=(10, 5))
plt.xlabel('Weather conditions')
plt.ylabel('Number of Incidents')
plt.show()
```

## Incidents by weather conditions
- help with understanding how weather affects aviation safety. During the VMC weather conditions is when most incidents occurred hence more caution and this allows a business or airline to enhance their risk mitigation startegies/ offer training to the staff.

```python
# plot a of bar graph incidents by purpose of flight
Aviation1['Purpose.of.flight'].value_counts().plot(kind='bar', title='Incidents by Purpose of Flight', figsize= (10,5))
plt.xlabel('Purpose of Flight')
plt.ylabel('Number of Incidents')
plt.show()
```

## Incidents by purpose of flight
- This graph shows different types of flights purposes and shows that personal/private and instructional flights had the highest incidents. It offers insights that the two should be observed carefully and specialized training programs should be offered, operationalized improvemnts and insurance services should be more targeted to the two.


```python
# incidents by country, histogram for top 10 countries
Aviation1['Country'].value_counts().head(10).plot(kind='bar', title='Incidents by Country', figsize=(10,5))
plt.xlabel('Country')
plt.ylabel('Number of Incidents')
plt.show()

```

## Incidents by country bar chart
- This is to visualize the average distribution of incidents across countries. understanding geographical incidents patterns reveals opportunities for regional safety programs, partnerships market expansion, eg. USA had the highest incidents compared to the rest of countries.

```python
# Histogram of distribution of total uninjured
Aviation1['Total.Uninjured'].plot(kind='hist', bins=5, title='Distribution of Total Uninjured People')
plt.xlabel('Total Uninjured')
plt.ylabel('Frequency')
plt.show()

```

## Distribution of total uninjured
- The histogram is to understand the frequency of incidents that result in less injuries/ zero inuries. A high number of incident with many uninjured passengers indicates safety features, effective procedures and resillient measures that reduce fatalities which is an attractive feature for marketing and emphasizes on business reliability. In this case there were more than 100 total uninjured in more than 80,000 incidents.

```python
# A bar graph of incidents by models
Aviation1['Model'].value_counts().head(10).plot(kind='bar', title='Incidents by Model', figsize=(10,5))
plt.xlabel('Model')
plt.ylabel('Number of Incidents')
plt.show()
```

### A bar graph of Incidents by models
- This is to identify which models are more prone to incidents/ accidents which informs investiments or operational focus. Also informs on which models with better safety records, lower insurance premiums, regulatory compliance advantages. In this case an investor should be careful when buying or investing in models like 152, 172, 172N, PA-28-140, 150 and 172M, for they have been involved in many accidents

```python
# A line graph of Trends of total fatal injuries overtime
Aviation1['Event.Date'] = pd.to_datetime(Aviation1['Event.Date'])
Aviation1.groupby(Aviation1['Event.Date'].dt.year)['Total.Fatal.Injuries'].sum().plot(kind='line', title='Trend of Fatal Injuries Over Time')
plt.xlabel('Year')
plt.ylabel('Total Fatal Injuries')
plt.show()
```

### A line graph of total fatal injuries overtime
- The graph is to analyze how safety has evolved over time. The downward trend in fatal injuries shows improvements in safety standards and technology making the industry more appealing for business.  

```python
# A correlation heatmap betwen injury types
plt.figure(figsize=(10,10))
sns.heatmap(Aviation1.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### A correlation heatmap between injury types
- The heatmap examines the relationship between different levels of the injuries. Understanding guides the design of better safety protocols/ onboard safety features that reduce injury severity.
